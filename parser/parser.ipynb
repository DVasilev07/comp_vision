```python
# \Получение ссылок и сохранение
site_mask = 'https://itboat.com/'
name = 0

for img in soup.find_all('img'):
    if type(img.get('src')) is str: # на всякий случай, проверка на то, что подается строка
        res_jpg = re.search(r'.jpg',img.get('src')) # картинки на сайте
        res_jpg_http = re.search(r'http[s]*\:\/\/[\w\W\s]+\.jpg', img.get('src')) # ссылки на картинки на сайте
    
    name += 1  
    name_str = str(name)
    # Блок по картинкам на сайте
    if res_jpg: # проверка, что нашли картинку
        src = img.get('src')
        if src: # проверка , что к картинке указан src
            #print(img.get('src'))
            adr = site_mask + img.get('src') # сцепляем с маской и получаем адрес
            response = http.request('GET', adr)
            out = open("./data/Positive/" + name_str + ".jpg" , 'wb')
            out.write(response.data)
            out.close()
            print(adr)
    # Блок по ссылкам на картинки на сайте        
    if res_jpg_http: # проверка, что нашли картинку
        src = img.get('src')
        if src: # проверка , что к картинке указан src
            print(img.get('src'))
```

```python
#### Цикл с записью
http = urllib3.PoolManager()
name = 0
for img in soup.find_all('img'):
    name += 1  
    name_str = str(name)
    #print(img.get('src'))
    res_jpg = re.search(r'http[s]*\:\/\/[\w\W\s]+\.jpg', img.get('src'))
    if res_jpg is not None:
        url = img.get('src')
        response = http.request('GET', url)
        out = open("./" + name_str + ".jpg" , 'wb')
        out.write(response.data)
        out.close()
        print(img.get('src'))
```
